```python
# üì¶ Install required packages (only needed once in Colab)
!pip install plotly openpyxl --quiet

# üìö Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline

# üìÇ Upload + Read file (Colab)
from google.colab import files

print("Please upload your Airbnb Excel data file (e.g. 1730285881-Airbnb_Open_Data.xlsx).")
uploaded = files.upload()
if not uploaded:
    print("No file uploaded ‚Äî will try fallback filename.")
# get uploaded or fallback
if uploaded:
    file_path = list(uploaded.keys())[0]
else:
    file_path = "1730285881-Airbnb_Open_Data.xlsx"
print(f"Using file: {file_path}")

# read dataframe
df = pd.read_excel(file_path, engine="openpyxl")

# --- Initial info ---
print("\n--- Initial dataset info ---")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Duplicates (before):", df.duplicated().sum())
display(df.head())

# --- Normalize columns & drop duplicates, useless cols ---
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r'\s+', '_', regex=True)
    .str.replace('-', '_', regex=False)
)

dup_before = df.duplicated().sum()
if dup_before > 0:
    df = df.drop_duplicates().reset_index(drop=True)
    print(f"Dropped {dup_before} duplicate rows.")
else:
    print("No duplicate rows.")

for c in ["house_rules", "license"]:
    if c in df.columns:
        df.drop(columns=[c], inplace=True)
        print(f"Dropped column {c}")

# --- Clean currency columns ---
def clean_currency_column(df, col):
    if col not in df.columns:
        return df
    df[col] = (
        df[col].astype(str)
        .str.replace(r'[\$,]', '', regex=True)
        .str.replace(r'\s+', '', regex=True)
        .replace({'': np.nan})
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")
    return df

df = clean_currency_column(df, "price")
# find service fee-like column
sf = [c for c in df.columns if "service" in c and "fee" in c]
service_fee_col = sf[0] if sf else None
if service_fee_col:
    df = clean_currency_column(df, service_fee_col)

# --- Fix datatypes ---
if "host_id" in df.columns:
    df["host_id"] = df["host_id"].astype(str).str.strip()
if "last_review" in df.columns:
    df["last_review"] = pd.to_datetime(df["last_review"], errors="coerce")
if "construction_year" in df.columns:
    df["construction_year"] = pd.to_numeric(df["construction_year"], errors="coerce")

# --- Missing values handling ---
rows_before = len(df)
rows_after_full_drop = df.dropna().shape[0]
print(f"\nRows before dropna: {rows_before}, after full dropna: {rows_after_full_drop}")

if rows_after_full_drop > max(1, int(rows_before * 0.2)):
    df = df.dropna().reset_index(drop=True)
    print("Dropped rows with any missing values.")
else:
    essentials = [c for c in ["price", "host_id", "neighbourhood", "neighbourhood_group"] if c in df.columns]
    if essentials:
        df = df.dropna(subset=essentials).reset_index(drop=True)
        print("Dropped rows missing essential columns:", essentials)
    else:
        print("Skipped dropna to avoid excessive data loss.")

print("Shape after NA handling:", df.shape)

# --- Clean neighbourhood_group & availability ---
if "neighbourhood_group" in df.columns:
    df["neighbourhood_group"] = (
        df["neighbourhood_group"].astype(str)
        .str.lower()
        .str.strip()
        .replace({"brookln": "brooklyn", "broooklyn": "brooklyn"})
        .str.title()
    )

if "availability_365" in df.columns:
    df["availability_365"] = pd.to_numeric(df["availability_365"], errors="coerce")
    df = df[df["availability_365"].between(0, 365)].reset_index(drop=True)

print("Final cleaned shape:", df.shape)
display(df.head())

# --- Helper to find columns ---
def find_col_containing(keyword):
    keyword = keyword.lower()
    for c in df.columns:
        if keyword in c:
            return c
    return None

property_col = find_col_containing("property_type")
neighbourhood_col = find_col_containing("neighbourhood")
neighbourhood_group_col = find_col_containing("neighbourhood_group")
host_id_col = find_col_containing("host_id")
reviews_per_month_col = find_col_containing("reviews_per_month")
identity_col = find_col_containing("host_identity_verified")
room_type_col = find_col_containing("room_type")
review_rate_col = None
for c in df.columns:
    if "review" in c and ("rate" in c or "rating" in c or "score" in c):
        review_rate_col = c
        break

print("\nResolved columns:")
print("property_type:", property_col)
print("neighbourhood:", neighbourhood_col)
print("neighbourhood_group:", neighbourhood_group_col)
print("price:", "price" if "price" in df.columns else None)
print("service_fee:", service_fee_col)
print("host_id:", host_id_col)
print("reviews_per_month:", reviews_per_month_col)
print("host_identity_verified:", identity_col)
print("room_type:", room_type_col)
print("review_rate:", review_rate_col)

# ====================
# EDA & Visuals
# ====================
sns.set(style="whitegrid")

# Q1
if property_col:
    print("\nQ1: Property Types")
    vc = df[property_col].value_counts()
    print(vc.head(10))
    plt.figure(figsize=(8,5))
    sns.countplot(y=property_col, data=df, order=vc.index)
    plt.title("Property Types Distribution")
    plt.tight_layout()
    plt.show()

# Q2
if neighbourhood_group_col:
    print("\nQ2: Listings by Neighbourhood Group")
    cnts = df[neighbourhood_group_col].value_counts().reset_index()
    cnts.columns = [neighbourhood_group_col, "count"]
    print(cnts)
    px.bar(cnts, x=neighbourhood_group_col, y="count", title="Listings per Neighbourhood Group").show()

# Q3
if neighbourhood_col and "price" in df.columns:
    print("\nQ3: Top 10 Neighbourhoods by Avg Price")
    avgp = df.groupby(neighbourhood_col)["price"].mean().sort_values(ascending=False).head(10)
    print(avgp)
    avgp.plot(kind="bar", figsize=(10,4), title="Top 10 Expensive Neighbourhoods")
    plt.tight_layout()
    plt.show()

# Q4
if "construction_year" in df.columns and "price" in df.columns:
    print("\nQ4: Construction Year vs Price")
    plt.figure(figsize=(8,5))
    sns.scatterplot(x="construction_year", y="price", data=df, alpha=0.5)
    plt.title("Construction Year vs Price")
    plt.tight_layout()
    plt.show()

# Q5
if host_id_col:
    print("\nQ5: Top 10 Hosts by Listing Count")
    top_hosts = df[host_id_col].value_counts().head(10)
    print(top_hosts)
    top_hosts.plot(kind="bar", figsize=(9,4), title="Top 10 Hosts by Listing Count")
    plt.tight_layout()
    plt.show()

# Q6
if reviews_per_month_col and identity_col:
    print("\nQ6: Verified Hosts vs Reviews per Month")
    df[identity_col] = (
        df[identity_col].astype(str)
        .str.lower()
        .str.strip()
        .map(lambda x: "yes" if x in ["true","t","yes","verified","1"] else "no")
    )
    plt.figure(figsize=(8,5))
    sns.boxplot(x=identity_col, y=reviews_per_month_col, data=df)
    plt.title("Verified vs Unverified Hosts ‚Äî Reviews per Month")
    plt.tight_layout()
    plt.show()

# Q7
if "price" in df.columns and service_fee_col:
    print("\nQ7: Price vs Service Fee")
    plt.figure(figsize=(8,5))
    sns.scatterplot(x="price", y=service_fee_col, data=df, alpha=0.5)
    plt.title("Price vs Service Fee")
    plt.tight_layout()
    plt.show()
    print("Correlation:", df[["price", service_fee_col]].corr().iloc[0,1])

# Q8
if review_rate_col and neighbourhood_group_col and room_type_col:
    print("\nQ8: Avg Review Rate by Group & Room Type")
    grouped = (
        df.groupby([neighbourhood_group_col, room_type_col])[review_rate_col]
        .mean()
        .reset_index()
        .rename(columns={review_rate_col: "avg_rating"})
    )
    display(grouped.head(20))
    plt.figure(figsize=(10,5))
    sns.barplot(x=neighbourhood_group_col, y="avg_rating", hue=room_type_col, data=grouped)
    plt.title("Avg Review Rate by Neighbourhood Group & Room Type")
    plt.tight_layout()
    plt.show()

# Q9
if host_id_col and "availability_365" in df.columns:
    print("\nQ9: Host Listings vs Availability")
    host_av = df.groupby(host_id_col)["availability_365"].mean().sort_values(ascending=False).head(10)
    print(host_av)
    host_av.plot(kind="bar", figsize=(9,4), title="Top Hosts by Avg Availability")
    plt.tight_layout()
    plt.show()

# ------------------------
# Final Summary Report
# ------------------------
print("\n\nüìä Summary of Insights\n")
print("1Ô∏è‚É£ Property Types:")
print("- Distribution of property types shown above.")

print("\n2Ô∏è‚É£ Neighbourhood Group with Most Listings:")
if neighbourhood_group_col:
    top_ng = df[neighbourhood_group_col].value_counts().idxmax()
    print(f"- The most frequent neighbourhood group is {top_ng}.")
else:
    print("- No neighbourhood_group data.")

print("\n3Ô∏è‚É£ Neighbourhoods with Highest Avg Price:")
if neighbourhood_col and "price" in df.columns:
    print("- Top neighbourhoods by avg price printed above.")
else:
    print("- No neighbourhood or price data.")

print("\n4Ô∏è‚É£ Construction Year vs Price:")
print("- The scatterplot helps see if newer properties ask higher prices. Trends depend on data.")

print("\n5Ô∏è‚É£ Top 10 Hosts by Listings:")
print("- Hosts with many listings are likely more commercial. See Q5 output above.")

print("\n6Ô∏è‚É£ Verified Hosts vs Reviews:")
print("- Verified identity often links with more consistent reviews (see boxplot).")

print("\n7Ô∏è‚É£ Price vs Service Fee:")
print("- Usually service fee increases with price. The scatter + correlation shows this.")

print("\n8Ô∏è‚É£ Average Review Rate (‚≠ê):")
print("- Differences in average ratings across neighbourhood groups & room types shown if data available.")

print("\n9Ô∏è‚É£ Host Listings vs Availability:")
print("- Hosts with many listings sometimes maintain higher availability. See Q9 output above.")


```
